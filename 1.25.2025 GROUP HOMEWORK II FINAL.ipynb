### Here is the Integrate and Fire Model for Homework II (OLD WORK)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
time_step = 0.1  # ms
total_time = 500  # ms
time = np.arange(0, total_time, time_step)
membrane_potential = np.full_like(time, -70.0)  # Start at resting potential (-70mV)
resting_potential = -70.0  # mV
threshold_potential = -55.0  # mV
reset_potential = -70.0  # mV (after spike)
spike_potential = 30.0  # mV (peak of action potential)
refractory_period = 5  # ms
refractory_time = int(refractory_period / time_step)
input_current = 15.0  # nA
membrane_resistance = 10.0  # M
membrane_capacitance = 100.0  # pF
tau = membrane_resistance * membrane_capacitance  # Time constant

# Simulation
spike_count = 0
for t in range(1, len(time)):
    if spike_count >= 5:
        break

    if membrane_potential[t-1] == spike_potential:
        # Apply refractory period
        membrane_potential[t:t+refractory_time] = reset_potential
        continue

    dV = (resting_potential - membrane_potential[t-1] + (membrane_resistance * input_current)) / tau
    membrane_potential[t] = membrane_potential[t-1] + dV * time_step

    # Check for spike
    if membrane_potential[t] >= threshold_potential:
        membrane_potential[t] = spike_potential  # Depolarization (spike)
        spike_count += 1

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, membrane_potential)
plt.axhline(y=threshold_potential, color='red', linestyle='--', label='Threshold (-55mV)')
plt.axhline(y=resting_potential, color='blue', linestyle='--', label='Resting Potential (-70mV)')
plt.title('Integrate-and-Fire Model Neuron Response (5 Spikes)')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
dt = 0.1  # Time step (ms)
T = 200  # Total time (ms)
time = np.arange(0, T, dt)

# LIF model parameters
V_rest = -70  # Resting potential (mV)
V_thresh = -55  # Threshold potential (mV)
V_reset = -70  # Reset potential after spike (mV)
R = 10  # Membrane resistance (M)
tau = 20  # Membrane time constant (ms)
I = 2  # Input current (nA) to generate 5 spikes
refractory_period = 5  # Refractory period (ms)

# Initialize variables
V = np.ones(len(time)) * V_rest
spikes = []
t_refractory = 0  # Tracks refractory period

# Simulate LIF neuron dynamics
for t in range(1, len(time)):
    if t_refractory > 0:  # Neuron is in refractory period
        V[t] = V_reset
        t_refractory -= dt
    else:
        dV = (-(V[t-1] - V_rest) + R * I) * (dt / tau)
        V[t] = V[t-1] + dV
        
        if V[t] >= V_thresh:  # Fire a spike
            V[t] = 0  # Artificial spike for visualization
            spikes.append(time[t])
            t_refractory = refractory_period  # Apply refractory period

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(time, V, label="Membrane Potential (mV)")
plt.axhline(y=V_thresh, color='r', linestyle='--', label="Threshold (-55mV)")
plt.axhline(y=V_rest, color='b', linestyle='--', label="Resting Potential (-70mV)")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.ylim(-100, 10)  # Extend y-axis down to -100 mV
plt.title("Leaky Integrate-and-Fire (LIF) Neuron Model")
plt.legend()
plt.show()
```

### Here is some information about our code:
#### Definition of LIF:
The Leaky Integrate and Fire model (LIF) refers to an updated version of the original Integrate and Fire (I&F) mathematical model. The I&F model was originally proposed to accurately describe neuronal firing patterns elicited in response to various stimuli. In the LIF model, leaky ion channels are incorporated to make the model more analogous to a biological neuron, as the presence of leaky channels models the membrane potential as changing non-linearly. For instance, the LIF models ability to replicate the presence of leaky ion channels leads to the membrane potential decaying more naturally over time, instead of resetting immediately as seen in the I&F model. Although the LIF model still integrates incoming stimuli and fires an action potential when a specific threshold is reached, the model requires an artificial spike generator to generate action potentials when threshold is reached, since it does not take into account the presence of voltage-gated ion channels that naturally occur in a biological neuron.

#### Definition of HH: 
The Hodgkin-Huxley (HH) model refers to the most refined version of the Integrate and Fire (I&F) mathematical model discussed thus far, and is most closely aligned to the functions of a neuron. Similar to the Leaky Integrate and Fire model (LIF), the HH model attempts to describe neuronal firing patterns elicited in response to various stimuli, while taking into account leaky ion channels. However, this model also incorporates voltage-gated ion channels, such as sodium (Na+) and potassium (K+) channels, making it more complex. Specifically, the HH model receives and integrates incoming stimuli. If the incoming stimuli cause the membrane potential to reach threshold, appropriate voltage-gated ion channels will open and/or close to represent the firing of an action potential. As a result, the model can simulate the nonlinear processes involving membrane potential dynamics that a biological neuron normally exhibits, without relying on the presence of an artificial spike generator to generate the action potential. However, one significant limitation of the HH model is that it assumes all ion channels present in the neuron behave in the same manner, and that the same properties can be seen across all aspects that define the neuronal membrane (dendrites, soma, and axons).

# Adding Inhibitory Inputs into LIF Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
dt = 0.1  # Time step (ms)
T = 200  # Total time (ms)
time = np.arange(0, T, dt)

# LIF model parameters
V_rest = -70  # Resting potential (mV)
V_thresh = -55  # Threshold potential (mV)
V_reset = -70  # Reset potential after spike (mV)
R = 10  # Membrane resistance (M)
tau = 20  # Membrane time constant (ms)
I_exc = 2  # Excitatory input current (nA) to generate spikes
refractory_period = 5  # Refractory period (ms)

# Inhibitory input parameters
I_inh_mean = 1  # Mean inhibitory current (nA)
I_inh_var = 0.5  # Variability in inhibitory current (nA)
I_inh = np.random.normal(I_inh_mean, I_inh_var, len(time))  # Time-varying inhibition

# Initialize variables for both cases
V_exc = np.ones(len(time)) * V_rest  # Excitation only
V_exc_inh = np.ones(len(time)) * V_rest  # Excitation + Inhibition
t_refractory_exc = 0  # Refractory period tracker (Exc only)
t_refractory_exc_inh = 0  # Refractory tracker (Exc + Inh)

spikes_exc = []  # Spike times for excitation only
spikes_exc_inh = []  # Spike times for excitation + inhibition

# Simulate LIF neuron dynamics
for t in range(1, len(time)):
    # Excitation only case
    if t_refractory_exc > 0:  # Refractory period active
        V_exc[t] = V_reset
        t_refractory_exc -= dt
    else:
        dV = (-(V_exc[t-1] - V_rest) + R * I_exc) * (dt / tau)
        V_exc[t] = V_exc[t-1] + dV
        if V_exc[t] >= V_thresh:  
            V_exc[t] = 0  # Artificial spike for visualization
            spikes_exc.append(time[t])
            t_refractory_exc = refractory_period

    # Excitation + Inhibition case
    if t_refractory_exc_inh > 0:  
        V_exc_inh[t] = V_reset
        t_refractory_exc_inh -= dt
    else:
        I_total = I_exc - I_inh[t]  # Total input with inhibition
        dV = (-(V_exc_inh[t-1] - V_rest) + R * I_total) * (dt / tau)
        V_exc_inh[t] = V_exc_inh[t-1] + dV
        if V_exc_inh[t] >= V_thresh:  
            V_exc_inh[t] = 0  
            spikes_exc_inh.append(time[t])
            t_refractory_exc_inh = refractory_period

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(time, V_exc, label="Excitatory Input Only", color='b')
plt.plot(time, V_exc_inh, label="Excitatory + Inhibitory Input", color='r', linestyle='dashed')
plt.axhline(y=V_thresh, color='k', linestyle='--', label="Threshold (-55mV)")
plt.axhline(y=V_rest, color='g', linestyle='--', label="Resting Potential (-70mV)")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.ylim(-100, 10)  
plt.title("LIF Neuron: Effect of Inhibition on Membrane Potential")
plt.legend()
plt.show()
```

### Simple neural network (open loop):

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def feedforward_nn(X, weights_input_hidden, weights_hidden_output):
    # Input to Hidden Layer
    hidden_input = np.dot(X, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)
    
    # Hidden to Output Layer
    final_input = np.dot(hidden_output, weights_hidden_output)
    final_output = sigmoid(final_input)
    
    return final_output

# Define input data (4 samples, 2 features each)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Randomly initialize weights
np.random.seed(42)
weights_input_hidden = np.random.randn(2, 3)  # 2 inputs, 3 hidden neurons
weights_hidden_output = np.random.randn(3, 1)  # 3 hidden neurons, 1 output neuron

# Compute output
output = feedforward_nn(X, weights_input_hidden, weights_hidden_output)

# Visualize results
plt.figure(figsize=(6, 4))
plt.scatter(range(len(output)), output, c='blue', label='Network Output')
plt.xticks(range(len(X)), ['(0,0)', '(0,1)', '(1,0)', '(1,1)'])
plt.xlabel('Input Pattern')
plt.ylabel('Output Value')
plt.title('Simple Feedforward Neural Network Output')
plt.legend()
plt.show()
```

## Topic Elaboration: Coincidence Detection

Coincidence detection is a phenomenon in which animals will use dual inputs converging where the location of convergence serves as an implication for location. This process allows for a neuron to fire an action potential when there are multiple excitatory inputs at the same time (which comes from sound waves entering the right ear and the left ear). Coincident detection allows the brain to determine sound localization, where the sound came from, as well as filter out noise, if the excitatory input does not reach the threshold (Stoll et al., 2023). It is important to note that this activity has to be detected within a short period of time to fire (Stoll et al., 2023). 

The ability for coincidence detection is said to be very important for survival, especially in the wild, to be able to rapidly and accurately determine where a sound is coming from (Campbell & King, 2004). For example, the red neuron responds best to a sound that is sourced from the left side, while the purple neuron responds best to a sound that was sourced from the right side. If a sound is coming from the center, it will reach the right and left ears at the same time and converge on the green neuron providing sound localization.
